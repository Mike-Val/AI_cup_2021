The method chosen to solve the tasks is the [Ant Colony Optimization](https://en.wikipedia.org/wiki/Ant_colony_optimization) algorithm with [Two opt heuristic](https://en.wikipedia.org/wiki/2-opt_algorithm) to improve the solutions produced at each iteration by the ants.

### Requirements
- C++17
- CMake >= 3.21

### Compilation

The solution was implemented in C++ and is supplied as source code which needs to be compiled.

The compilation is done using CMake with the following command:
```
cmake --build ./cmake-build-release --target AI_cup_2021
```

Which builds the binary file `AI_cup_2021` into the `./cmake-build-release/` directory.

Alternatively a `makefile` is also provided with some shortcuts:
- `make clean`: clean the build directory
- `make build`: build the project
- `make compile`: compile the code (recommended as it also calls the previous two commands)
- `make ${problem name}`: run the problem with the given name (Ex. `make ch130`)

In the `makefile` a variable seed with the same seed used to generate the results is also present.

### Usage

The program ```AI_cup_2021``` takes 2 required arguments and 1 optional argument:
- The seed to use for the random number generator (Ex. `1092841564`)
- The path to the problem to test (Ex. `./problems/ch130.tsp`)
- Optionally a path to a file where to save the solution found (Ex. `./results/tours/ch130.tsp`)

When running it outputs messages to the console. The first 3 lines are a summary of the problem that was loaded.<br>

Then it starts the solver and the timer for the 180 seconds of the time limit. <br>
Once the algorithm has converged to the best known solution or 180 seconds are about to elapse it stops and prints the time taken (in seconds), the distance of the solution found and the error (in percent).<br>

If an output file is supplied then also the name of the newly created file is printed.

### File output

The saved file (when an output path is supplied) is a `.tsp` file which looks like this:

### Repeatability

Since some problems are too big to fully converge to 0% error in the 3 minutes allowed, to reach a solution comparable to the one supplied in the excel file, some precautions need to be taken. The tests were performed under these conditions:

- The code was tested on a MacBook Pro (15-inch 2019) with 16 GB of 2400 MHz DDR4 RAM on macOS 11.6.1
- No other CPU intensive process was running at the same time of the tests. Programs like Mail or a web browser are fine as long as they are in background and are not performing tasks.
- The laptop was plugged in to charge and the airflow of the vents was not obstructed.

These requirements might seem harsh, but it was found that even small differences like being connected to power or not had a big impact on the result.

### Notes

The problem files are supplied in the `problems` directory. Using those is suggested, because of the way files are read.

For example the `rat783` given by the assignment has a blank space before the index of each set of coordinate which was removed manually, because it caused the code to read the coordinated incorrectly.

The path of any problem was no modified thus the results should still be regarded as correct.

## Results

Run all the following cells in order to see the results of the path found.

To note: the last edge of the tour is not drawn to show where the tour begins and ends.

````python
import numpy as np
import matplotlib.pyplot as plt
from typing import List, Tuple


def tsp_points(filename: str) -> List[Tuple[float, float]]:
    with open(filename, 'r') as f:
        lines = f.readlines()
        points = []
        for line in lines[7:-1]:
            idx, x, y = line.split()
            points.append([float(x), float(y)])
        return points


def plot_solution(problem: str, solution: str) -> None:
    points = tsp_points(problem)
    with open(solution, 'r') as f:
        lines = f.readlines()
        print(''.join(lines[:5]))
        path = [int(l.strip())-1 for l in lines[6:-1]]
    points = np.array([points[i] for i in path])
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot()
    ax.plot(points[:,0], points[:,1], marker='o')
    plt.show()
````

````python
plot_solution('../problems/ch130.tsp', 'tours/ch130.tsp')
````
