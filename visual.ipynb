```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
```

```python
class Problem:
    name = ''
    dimension = 0
    best_known = 0
    nodes = []
```

```python
def read_tsp_problem(filename):
	with open(filename) as f:
		lines = f.readlines()
	problem = Problem()
	for i, l in enumerate(lines):
		if l.startswith('NAME'):
			problem.name = l.split(':')[1].strip()
		elif l.startswith('DIMENSION'):
			problem.dimension = int(l.split(':')[1].strip())
		elif l.startswith('BEST_KNOWN'):
			problem.best_known = int(l.split(':')[1].strip())
		elif i > 6:
			problem.nodes.append(np.array(l.split()[1:], dtype=np.float64))
	return problem
```

```python
def plot_tour(P, tour):
    points = np.zeros((len(tour), 2))
    labels = []
    for i,p in enumerate(tour):
        points[i] = P.nodes[p]
        labels.append(str(p))
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot()
    ax.plot(points[:,0], points[:,1], marker='o')
    for i in range(len(tour)):
        ax.annotate(labels[i], (points[i,0], points[i,1]))
    plt.show()
```
